```python
import altair as alt
import pandas as pd
alt.renderers.enable('default')
```

```python
df = pd.read_excel('SubjectScores.xls')
df
```

```python
alt.Chart(df).mark_bar().encode(
    x='State',
    y='English',
    color='Science'
)
```

```python
interval = alt.selection_interval(encodings=['x'])

chart = alt.Chart(df).mark_square(size=50).encode(
    x='Mathematics',
    y='Science',
    color=alt.condition(interval, 'State', alt.value('lightgrey')),
    tooltip='State'
).properties(
    selection = interval,
    title = 'Subject wise student performance in percentages at Secondary level'
)

subjectScores = chart&chart.encode(
    x='English',
    y='MIL',
    tooltip='State'
).interactive()

subjectScores.save('subjectScores.html')
```

```python
interval = alt.selection_interval(encodings=['x'])

chart = alt.Chart(df).mark_point().encode(
    x='State',
    y='Science',
    color=alt.condition(interval, 'Mathematics', alt.value('lightgrey'))
).properties(
    selection = interval,
    title = 'Science Scores'
)

hist=alt.Chart(df).mark_bar().encode(
    x='MIL',
    y='State'
).transform_filter(
    interval
).properties (
    title = 'MIL Scores'
)


chart&hist
```

```python
(chart&hist).save('abc.html')
```

```python
# KGBV Enrollments

df_kg = pd.read_excel('KGBV Enrollment.xls')

df_kg ['GirlsPerColl'] = df_kg['No. of Girls Enrolled']/df_kg['No. of KGBVs functional under Samagra Shiksha']
df_kg.head()
```

```python
bar1 = alt.Chart(df_kg).mark_bar(size=15).encode(
    y = 'Name of State/UT',
    x = 'GirlsPerColl',
    color=alt.Color(value='lightgreen')
)

text=alt.Chart(df_kg).mark_text(dx=-3, dy=3).encode(
    y = 'Name of State/UT',
    x = 'GirlsPerColl',
    color=alt.Color(value='red'),
    text=alt.Text('GirlsPerColl',format='.1f')
)
chartKGBV = (bar1+text).properties (title = 'State/ UT wise Girls per KGBV ')
chartKGBV.save('chartKGBV.html')
# alt.layer(bar1, text, df=df_kg)
```

```python
# Facilities for CWSN

df_F = pd.read_excel('Facilities.xlsx')
df_F_8 = df_F[df_F['Inclusive Education']=='up to Highest Class VIII']
df_F_12 = df_F[df_F['Inclusive Education']=='Upto Highest Class - XII']
df_F_10 = df_F[df_F['Inclusive Education']=='Upto Highest Class X or XII']

df_F_10
```

```python
provisionSpecialNeed = alt.Chart(df_F_8).mark_square().encode(
    y = 'Inclusive Education',
    x='Category',
    size='Physical Count',
#     color='Financial (Rs in lakh)',
    tooltip = 'Financial (Rs in lakh)'
) + alt.Chart(df_F_10).mark_square().encode(
        y = 'Inclusive Education',
        x='Category',
        size='Physical Count',
        tooltip = 'Financial (Rs in lakh)'
    ) + alt.Chart(df_F_12).mark_square().encode(
        y = 'Inclusive Education',
        x='Category',
        size='Physical Count',
        tooltip = 'Financial (Rs in lakh)'
    ).properties (
    title = 'Provisions for Children with Special Needs Under Samagra Siksha 2018-19'
)

provisionSpecialNeed.save('provisionSpecialNeed.html')
```
